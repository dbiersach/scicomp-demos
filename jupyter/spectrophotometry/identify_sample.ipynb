```python
# identify_sample.ipynb

import os

import numpy as np


def calc_rel_error(obs, exp):
    err = 0.0
    for i in range(len(obs)):
        err += abs(obs[i] - exp[i]) / exp[i]
    return err


# Read unknown sample data file
sample_name = "Unknown"
folder_path = os.path.dirname(os.path.realpath("__file__"))
file_path = os.path.join(folder_path, sample_name + ".csv")
unknown_levels = np.genfromtxt(file_path, delimiter=",")

# Read known sample files
folder_path = os.path.join(folder_path, "samples")
sample_files = os.listdir(folder_path)

# Create empty lists to hold comparison value for each known sample
sample_names = []
scores = []

# Find relative error between each known sample's wavelengths and the unknown sample
for file in sample_files:
    file_path = os.path.join(folder_path, file)
    sample_levels = np.genfromtxt(file_path, delimiter=",")
    sample_names.append(file[:-4])  # Remove ".csv" from filename
    scores.append(calc_rel_error(unknown_levels, sample_levels))

# Display the known sample having the least difference to the unknown sample
match_index = scores.index(min(scores))
print(f"Best match: {sample_names[match_index]}")
```
