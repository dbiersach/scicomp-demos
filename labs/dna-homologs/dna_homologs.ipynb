```python
# dna_homologs.ipynb

# Cell 01 - Find Longest Repeating Substring

def match(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1
    s3 = ""
    for i in range(len(s1)):
        if s1[i] == s2[i]:
            s3 += s1[i]
        else:
            break
    return s3


def lrss(seq):
    seq = seq.upper()

    # Step 1 - Form the suffixes list
    suffixes = []
    for i in range(len(seq)):
        suffixes.append(seq[i:])

    # Step 2 - Sort the suffixes list
    suffixes.sort()

    # Step 3 - Scan the suffixes list
    longest = ""
    for i in range(len(suffixes) - 1):
        s1 = suffixes[i]
        s2 = suffixes[i + 1]
        candidate = match(s1, s2)
        if len(candidate) > len(longest):
            longest = candidate

    return longest


dna_sequence = "AACAAGTTTACAAGC"
print("The longest recurring substring in the sequence:")
print(dna_sequence)
print(f"is {lrss(dna_sequence)}")
```

```python
# Cell 02 - Find LRSS in a Fruit Fly Gene

import os

dna_sequence_file = "fruitfly.txt"
folder_path = os.path.dirname(os.path.realpath("__file__"))
file_path = os.path.join(folder_path, dna_sequence_file)
dna_sequence = open(file_path, "r").read()

print("The longest recurring substring in the Fruit Fly gene")
print(f"is {lrss(dna_sequence)}")
```
