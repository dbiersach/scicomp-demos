```python
# continued_fractions.ipynb
# Cell 1

import math

from IPython.display import Math


def normalize_cf(cf):
    if len(cf) > 2:
        if cf[-1] == 1 and cf[-2] != 1:
            cf[-2] += 1
            cf.pop(-1)
    return cf


def encode_cf(x, numterms=20):
    cf = []
    while len(cf) < numterms:
        cf.append(math.floor(x))
        x = x - math.floor(x)
        if x < 1e-11:
            break
        x = 1 / x
    return normalize_cf(cf)


def decode_cf(cf):
    hn, kn = 0, 0
    b_1, h_1, k_1 = 1, 1, 0
    h_2, k_2 = 0, 1
    for term in cf:
        an, bn = term, 1
        hn = an * h_1 + b_1 * h_2
        kn = an * k_1 + b_1 * k_2
        b_1 = bn
        h_1, h_2 = hn, h_1
        k_1, k_2 = kn, k_1
    return hn, kn


def print_cf(x):
    cf = encode_cf(x)
    print(f"The value {x} is encoded as {cf}")
    hn, kn = decode_cf(cf)
    x = hn / kn
    print(f"The cf {cf} is decoded as the value {x}")
    display(Math(rf"\frac{{{hn:,}}}{{{kn:,}}}\approx{{{x}}}"))
    print()


print_cf(3.245)

print_cf(0.825)
```

```python
# Cell 5


def decode_gencf(coeffs):
    a0, b0, Ai, Bi, Ci, Di, Ei = coeffs
    an, bn = a0, b0
    hn, kn = 0, 0
    b_1, h_1, k_1 = 1, 1, 0
    h_2, k_2 = 0, 1
    for n in range(1, 20):
        hn = an * h_1 + b_1 * h_2
        kn = an * k_1 + b_1 * k_2
        b_1 = bn
        h_1, h_2 = hn, h_1
        k_1, k_2 = kn, k_1
        an = Di * n + Ei
        bn = Ai * n * n + Bi * n + Ci
    return hn / kn


def print_gencf(label, coeffs):
    x = decode_gencf(coeffs)
    print(f"{label} Generalized Continued Fraction")
    print(f"Est. Pi     : {x:.15f}")
    print(f"Act. Pi     : {math.pi}")
    print(f"Abs. % Err  : {abs((math.pi - x)/math.pi):.14%}")


x = print_gencf("Euler's", (3, 1, 4, 4, 1, 0, 6))
```
