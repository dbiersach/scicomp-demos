```python
# welcome.ipynb

# Welcome to the Scientific Computing Demonstrations
# Labs developed by Dave Biersach (dbiersach@bnl.gov)

%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 500)
f_top = np.sqrt(1 - (np.abs(x) - 1) ** 2)
f_bot = np.arccos(1 - np.abs(x)) - np.pi

ax = plt.gca()
ax.plot(x, f_top, color="red")
ax.plot(x, f_bot, color="red")

ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 1.5)

ax.set_title("Welcome to Scientific Computing Demonstrations!")
ax.set_aspect("equal")

plt.show()
```
