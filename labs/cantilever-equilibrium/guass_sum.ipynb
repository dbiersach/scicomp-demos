```python
# gauss_sum.ipynb

import time

# Cell 01 - Sum natural numbers by looping

n = 100_000_000

print(f"Summing the first {n:,} natural numbers...")

start_time = time.time()

sum = 0
for k in range(n + 1):
    sum = sum + k

end_time = time.time()

print(f"Total by looping = {int(sum):,}")

print(f"Total run time = {end_time - start_time:.3f} seconds")
```

```python
# Cell 02 - Sum natural numbers by Gaussian Summation

print(f"Summing the first {n:,} natural numbers using Gauss...")

start_time = time.time()

sum = n / 2 * (n + 1)

end_time = time.time()

print(f"Total by Gaussian Summation = {int(sum):,}")

print(f"Total run time = {end_time - start_time:.3f} seconds")
```
