```python
# gauss_sum.ipynb

# Cell 01 - Sum the first 1000 natural numbers

n = 1000

sum_by_looping = 0
for k in range(n + 1):
    sum_by_looping = sum_by_looping + k

sum_by_gauss = n / 2 * (n + 1)

print(f"Sum of the first {n:,} natural numbers:")
print(f"...by looping over every value = {sum_by_looping:,}")
print(f"...by direct Gauss Summation   = {sum_by_gauss:,.0f}")
```
