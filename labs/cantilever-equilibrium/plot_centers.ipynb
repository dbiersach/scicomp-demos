```python
# plot_centers.ipynb

%matplotlib widget

import matplotlib.pyplot as plt
from blocks import BlockList

# Cell 01 - Plot center-of-mass in X & Y dimension
# for 14-block (7 ensemble) Jenga pile

# Calculate coordinates of center of mass after
# placing each ensemble in a 14-block Jenga pile
x_14_pile, y_14_pile = [0], [0]
block_list = BlockList()
for _ in range(1, 8):
    block_list.place_block(7.5, 1.5)
    block_list.place_block(1.5, 10.5)
    x_14_pile.append(block_list.center_x)
    y_14_pile.append(block_list.center_y)
    block_list.move_blocks(3, 3)

fig = plt.figure()
fig.suptitle("Center of Mass Per Ensemble (14 block pile)")
ax = plt.gca()
ax.plot(range(8), x_14_pile, label=f"x-Axis Center of Mass")
ax.plot(range(8), y_14_pile, label=f"y-Axis Center of Mass")
ax.set_xlabel("Number of Ensembles")
ax.set_ylabel("Location (cm)")
ax.legend()
ax.set_xlim(1)
plt.show()
```

```python
# Cell 02 - Plot center-of-mass in X & Y dimension
# for 15-block (7 ensemble, 3 starting blocks) Jenga pile

# Calculate coordinates of center of mass after
# placing each ensemble in a 15-block Jenga pile
x_15_pile, y_15_pile = [0], [0]
# Don't forget to place the extra initial block in
# the first ensemble for a 15-block Jenga pile
block_list = BlockList()
block_list.place_block(10.5, 4.5)
for n in range(1, 8):
    block_list.place_block(7.5, 1.5)
    block_list.place_block(1.5, 10.5)
    x_15_pile.append(block_list.center_x)
    y_15_pile.append(block_list.center_y)
    block_list.move_blocks(3, 3)

fig = plt.figure()
fig.suptitle("Center of Mass Per Ensemble (15 block pile)")
ax = plt.gca()
ax.plot(range(8), x_15_pile, label=f"x-Axis Center of Mass")
ax.plot(range(8), y_15_pile, label=f"y-Axis Center of Mass")
ax.set_xlabel("Number of Ensembles")
ax.set_ylabel("Location (cm)")
ax.legend()
ax.set_xlim(1)
plt.show()
```
