```python
# jenga-14.ipynb

# Cell 01 - Place first ensemble (2 blocks)

from blocks import BlockList

# Create an empty pile
block_list = BlockList()

# Place 1st horizontal block
block_list.place_block(7.5, 1.5)

# Place 1st vertical block
# Note: This block stands on top of 1st horizontal block!
block_list.place_block(1.5, 10.5)

# Print center of mass of pile
block_list.print_center()
```

```python
# Cell 02 - Place first two ensembles (4 blocks)

block_list = BlockList()

# Place ensemble #1
block_list.place_block(7.5, 1.5)
block_list.place_block(1.5, 10.5)

# Make space to add another ensemble at the origin
block_list.move_blocks(3, 3)

# Place ensemble #2
block_list.place_block(7.5, 1.5)
block_list.place_block(1.5, 10.5)

block_list.print_center()
```

```python
# Cell 03 - Construct largest pile of 2-block ensembles that is stable

block_list = BlockList()

block_list.place_block(7.5, 1.5)
block_list.place_block(1.5, 10.5)

while block_list.center_x < 15.0:
    block_list.print_center()
    block_list.move_blocks(3, 3)
    block_list.place_block(7.5, 1.5)
    block_list.place_block(1.5, 10.5)
```
